## Plotting multiple graphs using a common axis

```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')

t=[1,2,3,4]
temperature=[100,200,300,400]
dewpoint=[124,222,443,332]

plt.axes([1/5,2/5,3/5,4/5])
plt.plot(t,temperature,'red')
plt.plot(t,dewpoint,'blue')
plt.show()
```

### plt.axes()

```python
plt.axes([1/5,2/5,3/5,4/5])
plt.plot(t,temperature,'red')
plt.xlabel('x')
plt.title('red')
plt.show()
plt.axes([1/5,2/5,3/5,4/5])
plt.plot(t,dewpoint,'blue')
plt.xlabel('x1')
plt.title('blue')
plt.show()
```

### plt.subplot()

```python
plt.subplot(2,1,1) #no of rows, no of columns, and which subplot no to activate
#subplot number is indexed from no 1, row-wise from top left
plt.plot(t,temperature,'red')
plt.xlabel('x')
plt.title('red')
plt.subplot(2,1,2)
plt.plot(t,dewpoint,'blue')
plt.xlabel('x1')
plt.title('blue')
```

## Creating meshgrid

### np.linspace()

```python
import numpy as np
u=np.linspace(-2,2,200) #start #end #no of values generated in between
v=np.linspace(-2,2,200) #start #end #no of values generated in between
X,y=np.meshgrid(u,v)
Z=X**2+y**2
from matplotlib import pyplot as plt
plt.set_cmap('gray') #setting the color
plt.pcolor(Z) #setting the mesh color will be pseudo color if cmap not set
plt.show()
plt.savefig('test.png')
```

### Combining

```python
import numpy as np
from matplotlib import pyplot as plt
u=np.linspace(-2,2,200) #start #end #no of values generated in between
v=np.linspace(-2,2,200) #start #end #no of values generated in between
X,Y=np.meshgrid(u,v)
Z=X**2+y**2

# Generate a default contour map of the array Z
plt.subplot(2,2,1)
plt.contour(X,Y,Z)

# Generate a contour map with 20 contours
plt.subplot(2,2,2)
plt.contour(X,Y,Z,20)

# Generate a default filled contour map of the array Z
plt.subplot(2,2,3)
plt.contourf(X,Y,Z)
# Generate a default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20)

# Improve the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

## Example2

```python
# Create a filled contour plot with a color map of 'viridis'
plt.subplot(2,2,1)
plt.contourf(X,Y,Z,20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

# Create a filled contour plot with a color map of 'gray'
plt.subplot(2,2,2)
plt.contourf(X,Y,Z,20,  cmap='gray')
plt.colorbar()
plt.title('Gray')

# Create a filled contour plot with a color map of 'autumn'
plt.subplot(2,2,3)
plt.contourf(X,Y,Z,20,  cmap='autumn')
plt.colorbar()
plt.title('Autumn')

# Create a filled contour plot with a color map of 'winter'
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20,  cmap='winter')
plt.colorbar()
plt.title('Winter')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```

### 2D histograms with rectangles and hexagons

```python
#2D
import numpy as np
array1=np.random.rand(200,)
array2=np.random.rand(200,)
plt.hist2d(array1,array2,bins=(10,20) ) #horizontal array #vertical array @no of rectangles to draw in the horizontal direction #no to draw in vertical
plt.set_cmap('viridis') 
plt.colorbar()
plt.show()
```

You can specify the rectangular region in which the samples are counted in constructing the 2D histogram. The optional parameter required is range=((xmin, xmax), (ymin, ymax)) where
xmin and xmax are the respective lower and upper limits for the variables on the x-axis and
ymin and ymax are the respective lower and upper limits for the variables on the y-axis. Notice that the optional range argument can use nested tuples or lists.

Hexbin

```python
#2D
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
array1=np.random.rand(400,)
array2=np.random.rand(400,)
plt.hexbin(array1,array2,gridsize=(20,10) ) #for gridsize, first arguemtn is no of horizontal heagon. seocnd is no in vertical direction
plt.set_cmap('viridis') 
plt.colorbar()
plt.show()
```

extent is similar to range in 2D bins.

## Visualising with Seaborn

```python
import pandas as pd
import numpy as np
auto=pd.read_csv('https://raw.githubusercontent.com/shifteight/R/master/ISLR/Auto.csv')
auto['horsepower'] = auto.horsepower.apply(lambda x: x.replace('?', '0'))
auto['horsepower'] = auto.horsepower.apply(lambda x: int(x))
print(auto.keys())
```

### Simple linear regression 

```python
# Import plotting modules
from matplotlib import pyplot as plt
import seaborn as sns

# Plot a linear regression between 'weight' and 'hp'
sns.lmplot(x='weight', y='horsepower', data=auto)

# Display the plot
plt.show()
```

### Residual of regression

```python
#### Import plotting modules
from matplotlib import pyplot as plt
import seaborn as sns

#### Generate a green residual plot of the regression between 'hp' and 'mpg'
sns.residplot(x='horsepower', y='mpg', data=auto, color='green')

#### Display the plot
plt.show()
```

### Higher-order regressions

```python
#### Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')

#### Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, scatter=None, color='blue', label='order 1')

#### Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, scatter=None,order=2, color='green', label='order 2')

#### Add a legend and display the plot
plt.legend(loc='upper right')
plt.show()
```

## Grouping linear regresisons by hue

```python
#### Plot a linear regression between 'weight' and 'hp', with a hue of 'origin' and palette of 'Set1'
sns.lmplot(x='weight',y='horsepower',data=auto,palette='Set1',hue='origin')

#### Display the plot
plt.show()
```

## Grouping linear regressions by row or column

```python
#### Plot linear regressions between 'weight' and 'hp' grouped row-wise by 'origin'
sns.lmplot(x='weight',y='horsepower',data=auto,palette='Set1',row='origin')
    

#### Display the plot
plt.show()
```

### Visualising univariate distribution


```python
import seaborn as sns
import numpy as np
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/tips.csv')

plt.subplot(2,2,1)
sns.stripplot(x='day',y='tip',data=df)

plt.subplot(2,2,2)
sns.stripplot(x='day',y='tip',data=df,jitter=True)

plt.subplot(2,2,3)
sns.swarmplot(x='day',y='tip',data=df)

plt.subplot(2,2,4)
sns.swarmplot(x='day',y='tip',data=df,hue='sex')
plt.show()

print(df.head())
print(type(df.day))
print(type(df.tip))
```

```python
sns.swarmplot(y='day',x='tip',data=df,hue='sex',orient='h')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=df)
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=df)
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=df,jitter=True)
sns.violinplot(x='day',y='tip',data=df,color='lightgray')
plt.show()
```

## Example 1

```python
# Make a strip plot of 'hp' grouped by 'cyl'
plt.subplot(2,1,1)
sns.stripplot(x='cylinders', y='horsepower', data=auto)

# Make the strip plot again using jitter and a smaller point size
plt.subplot(2,1,2)
sns.stripplot(x='cylinders', y='horsepower', data=auto, jitter=True, size=3)

# Display the plot
plt.show()
```

## Example 2

Another grouping can be added in using the hue keyword. For instance, using sns.swarmplot(x='type', y='length', data=df, hue='build year') makes a swarm plot from the DataFrame df with the 'length' column values spread out vertically, horizontally grouped by the column 'type' and each point colored by the categorical column 'build year'.

```python
# Generate a swarm plot of 'hp' grouped horizontally by 'cyl'  
plt.subplot(2,1,1)
sns.swarmplot(x='cylinders',y='horsepower',data=auto)

# Generate a swarm plot of 'hp' grouped vertically by 'cyl' with a hue of 'origin'
plt.subplot(2,1,2)
sns.swarmplot(x='horsepower',y='cylinders',hue='origin',data=auto, orient='h')

# Display the plot
plt.show()
```

## Example 3

```python
# Generate a violin plot of 'hp' grouped horizontally by 'cyl'
plt.subplot(2,1,1)
sns.violinplot(x='cylinders', y='horsepower', data=auto)

# Generate the same violin plot again with a color of 'lightgray' and without inner annotations
plt.subplot(2,1,2)
sns.violinplot(x='cylinders', y='horsepower', data=auto,color='lightgray',inner=None)
# Overlay a strip plot on the violin plot
sns.stripplot(x='cylinders', y='horsepower', data=auto,size=1.5,jitter=True)

# Display the plot
plt.show()
```
