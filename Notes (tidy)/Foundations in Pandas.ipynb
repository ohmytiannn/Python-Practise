## Zip(key_list,value_list)

Takes two lists: the first being the the key one and the second one being the value list
The value list is a list of list
pd.DataFrame(dict(list(zip(key_list,value_list))))
###use the following code so that you will be able to build a dataframe from list

```python
# Zip the 2 lists together into one list of (key,value) tuples: zipped
zipped = list(zip(list_keys,list_values))

# Inspect the list using print()
print(zipped)

# Build a dictionary with the zipped list: data
data = dict(zipped)

# Build and inspect a DataFrame from the dictionary: df
df = pd.DataFrame(data) 
print(df)
```
